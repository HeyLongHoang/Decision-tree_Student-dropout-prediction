```python
def split(df, ratio):
    new_df = df.sample(frac = 1)
    idx = int(len(new_df) * ratio)
    X_train = new_df.iloc[:idx, :-1]
    y_train = new_df.iloc[:idx, -1]
    X_test = new_df.iloc[idx:, :-1]
    y_test = new_df.iloc[idx:, -1]
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = split(df, ratio=0.995)

X_train.to_csv('data/train.csv')
y_train.to_csv('data/y_train.csv')
X_test.to_csv('data/test.csv')
y_test.to_csv('data/y_test.csv')
```
